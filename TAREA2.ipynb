```python
import os, geopandas as gpd
#Mapa 1: países
paises=gpd.read_file(os.path.join("data","World_Countries","World_Countries.shp"))
```

```python
#Exploración básica de la data
paises.columns
```

```python
paises.head()
```

```python
paises.info()
```

```python
# Filtrar país Argentina
Argentina=paises[paises.COUNTRY=='Argentina']
```

```python
#1 Mapa Argentina
Argentina.plot(facecolor="lightyellow", edgecolor='black',
              linewidth=0.5)
```

```python
# Data 2: rios
rios=gpd.read_file(os.path.join("data","World_Hydrography","World_Hydrography.shp"))
```

```python
#Mapa de los rios
rios.plot(edgecolor='blue',
          linewidth=1)
```

```python
#Rios solo en Argentina (clipped)
Arg_rios=gpd.clip(gdf=rios,mask=Argentina)
```

```python
#Mapa rios en Argentina
Arg_rios.plot(edgecolor='blue',
              linewidth=1)
```

```python
#Data 3: Departamenos Argentina (Divisiones Territoriales de segundo orden)
Arg_departamentos=gpd.read_file(os.path.join("data","fronteras","gadm41_ARG_2.shp"))
#Link data: https://gadm.org/download_country.html#google_vignette
```

```python
#Mapa departamenos de Argentina (Divisiones Territoriales de segundo orden)
Arg_departamentos.plot(edgecolor='red',facecolor="lightyellow",
                       linewidth=0.5)
```

```python
#Verificar que tengan el mismo CRS
Argentina.crs==Arg_rios.crs==Arg_departamentos.crs
```

```python
# Mapa conjunto (Argentina+ rios + departamentos)
base = Argentina.plot(edgecolor='black', linewidth=0.4,figsize=(6,6))
Arg_rios.plot(edgecolor='blue',linewidth=1.5,
              ax=base)
Arg_departamentos.plot(facecolor="lightyellow",edgecolor='red', linewidth=0.5,
                     ax=base)
```

```python
#Proyección
#Verificando las unidades
Argentina.crs.axis_info
```

```python
import warnings
warnings.filterwarnings('ignore')

Argentina.centroid
#Se comprueba que la data no está proyectada!!
```

```python
#Ajustando la proyección
#Link proyección: https://epsg.io/?q=argentina+kind%3APROJCRS
Argentina.to_crs(9252).crs.axis_info
```

```python
Argentina.to_crs(9252).centroid
#La data ya está proyectada
```

```python
#Centroide en la data proyectada
base9252=Argentina.to_crs(9252).plot()
Argentina.to_crs(9252).centroid.plot(color="orange",ax=base9252)
```

```python
# Proyectar todos los mapas
Argentina_9252=Argentina.to_crs(9252)
Arg_rios_9252=Arg_rios.to_crs(Argentina_9252.crs)
Arg_departamentos_9252=Arg_departamentos.to_crs(Argentina_9252.crs)
```

```python
# Data 4: aeropuertos
import pandas as pd
aeropuertos=pd.read_csv(os.path.join("data","airports.csv"))
```

```python
aeropuertos.columns
```

```python
# Contiene aeropuertos de todos los países
aeropuertos.head()
```

```python
# Filtrando solo aeropuertos de Argentina
Arg_aeropuertos=aeropuertos[aeropuertos['iso_country']=='AR']
```

```python
Arg_aeropuertos.head()
```

```python
#Mantenemos columnas de interés
mantener_col=['name','type','latitude_deg','longitude_deg','elevation_ft','iso_region','municipality']
Arg_aeropuertos=Arg_aeropuertos[mantener_col]
```

```python
Arg_aeropuertos.info()
```

```python
Arg_aeropuertos.head()
```

```python
#No se grafica bien porque la data de aeropuertos no está proyectada (está en grados - degrees)
base=Argentina_9252.plot(color='lightyellow', edgecolor='black', linewidth=0.4)
Arg_aeropuertos.plot.scatter(x='longitude_deg', y='latitude_deg', ax=base)
```

```python
#Convertimos Arg_aeropuertos en un geodata frame
Arg_aeropuertos2=gpd.GeoDataFrame(data=Arg_aeropuertos.copy(),
                 geometry=gpd.points_from_xy(Arg_aeropuertos.longitude_deg,
                                             Arg_aeropuertos.latitude_deg),
                 crs=Argentina.crs.to_epsg())
```

```python
base=Argentina_9252.plot(color='lightyellow', edgecolor='black', linewidth=0.4)
Arg_aeropuertos2.plot(ax=base)
```

```python
#Proyectamos la data de aeropuertos
Arg_aeropuertos_9252=Arg_aeropuertos2.to_crs(9252)
```

```python
# Gráfico aeropuertos de argentina
base=Argentina_9252.plot(color='lightyellow', edgecolor='black', linewidth=0.4)
Arg_aeropuertos_9252.plot(ax=base, markersize=2, color='purple')
```

```python
#Veamos los tipos de aeropuertos del mapa
Arg_aeropuertos_9252['type'].value_counts()
```

```python
# Mapa conjunto (Argentina + rios + departamentos + aeropuertos) con la misma proyección
base = Argentina_9252.plot(edgecolor='black', color='lightyellow', linewidth=0.4,figsize=(6,6))
Arg_rios_9252.plot(edgecolor='blue',linewidth=2,
              ax=base)
Arg_departamentos_9252.plot(facecolor="lightyellow",edgecolor='red', linewidth=0.5,
                     ax=base)
Arg_aeropuertos_9252.plot(markersize=2.5, color='purple',
                    ax=base)
```

```python
#Grabamos el mapa
import os

Argentina_9252.to_file(os.path.join("maps","ArgentinaMapa_9252.gpkg"),layer='pais', driver="GPKG")
Arg_rios_9252.to_file(os.path.join("maps","ArgentinaMapa_9252.gpkg"),layer='rios', driver="GPKG")
Arg_departamentos_9252.to_file(os.path.join("maps","ArgentinaMapa_9252.gpkg"),layer='departamentos', driver="GPKG")
Arg_aeropuertos_9252.to_file(os.path.join("maps","ArgentinaMapa_9252.gpkg"),layer='aeropuertos', driver="GPKG")
```
